```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
df=pd.read_csv("/content/country_vaccinations.csv")
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df
```

```python
#fill NAN with 0
df.fillna(0,inplace=True)
df.drop(df.index[df['iso_code']==0], inplace=True)
```

```python
df.isnull().sum()
```

```python
#describe is also used
df.info()
```

```python
df['date']=pd.to_datetime(df['date'],format='%Y-%m-%d')
```

```python
df.columns
```

```python
df.drop(['people_fully_vaccinated','daily_vaccinations_raw','people_fully_vaccinated_per_hundred','daily_vaccinations_per_million', 'people_vaccinated_per_hundred'
,'source_name','source_website'],axis=1,inplace=True)
```

```python
df_USA=df[df["iso_code"]=="USA"].copy()
df_USA
```

```python
#drop dates with missing values
df_USA.drop(df_USA.index[df_USA['total_vaccinations']==0],inplace=True)
```

```python
plt.figure(figsize=(18,6))
sns.lineplot(data=df_USA,x="date",y="total_vaccinations")
plt.title("Total Vaccination in USA")
plt.xticks(rotation=45)
plt.show()
```

```python
#plot daily vaccination as function of date
plt.figure(figsize=(18,6))
sns.lineplot(data=df_USA,x="date",y="daily_vaccinations")
plt.xticks(rotation=90)
plt.title("Daily Vaccination in USA")
plt.show()
```

```python
#plot for INDIA
```

```python
df_IND=df[df["iso_code"]=="IND"].copy()
df_IND
```

```python
#drop dates with missing values
df_IND.drop(df_IND.index[df_IND['total_vaccinations']==0],inplace=True)
```

```python
plt.figure(figsize=(18,6))
sns.lineplot(data=df_IND,x="date",y="total_vaccinations")
plt.title("Total Vaccination in INDIA")
plt.xticks(rotation=45)
plt.show()
```

```python
#plot daily vaccination as function of date
plt.figure(figsize=(18,6))
sns.lineplot(data=df_IND,x="date",y="daily_vaccinations")
plt.xticks(rotation=90)
plt.title("Daily Vaccination in INDIA")
plt.show()
```

```python
#Group by total vaccination given by the country and sort decending to indentify the top 10 countries
vacc_by_country=df.groupby('country').max().sort_values('total_vaccinations', ascending=False)
vacc_by_country=vacc_by_country.iloc[:10]
vacc_by_country
```

```python
#now sort by total vaccinations per 100
vacc_by_country=vacc_by_country.sort_values('total_vaccinations_per_hundred',ascending=False)
vacc_by_country
```

```python
plt.figure(figsize=(18,6))
plt.bar(vacc_by_country.index,vacc_by_country.total_vaccinations_per_hundred)

plt.xticks(rotation=90)
plt.ylabel('Vaccinations per 100')
plt.xlabel('Country')
plt.show()
```

```python
#sort by the total vaccinations deliverd by country and group by countried
```

```python
total_vacc_by_country=df.groupby('country').max().sort_values('total_vaccinations', ascending=False)
total_vacc_by_country=total_vacc_by_country.iloc[:10]
total_vacc_by_country
```

```python
plt.figure(figsize=(16,7))
plt.bar(total_vacc_by_country.index,total_vacc_by_country.total_vaccinations)

plt.xticks(rotation=90)
plt.ylabel('Vaccinations per Country')
plt.xlabel('Country')
plt.show()
```

```python
#sort the total vaccination delivered by the country group by vaccines
vacc_names_by_country=df.groupby('vaccines').max().sort_values('total_vaccinations', ascending=False)
vacc_names_by_country.head()

```

```python
vacc_names_by_country=vacc_names_by_country.iloc[:10]
vacc_names_by_country
```

```python
#Reset the index to move vaccines from being index to a column
#so that this easy to  plot using seaborn
vacc_names_by_country=vacc_names_by_country.reset_index()
vacc_names_by_country
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=vacc_names_by_country,x='vaccines',y='total_vaccinations',hue='country',dodge=False)
plt.xticks(rotation=90)
plt.show()
```
